```python
from collections import Counter

def replaced_params_count(replaced_params_list):
    """
    Counts occurrences of values in the 'values' lists across multiple replaced_params dictionaries.

    :param replaced_params_list: List of replaced_params dictionaries.
    :return: Counter with the count of each value.
    """
    count = Counter()
    for replaced_params in replaced_params_list:
        for values in replaced_params.values():
            count.update(values)
    return count

# Example usage:
replaced_params_list = [
    {"layer1": [1, 2, 3], "layer2": [2, 3, 4]},
    {"layer1": [3, 4, 5], "layer3": [1, 5, 6]}
]

result = replaced_params_count(replaced_params_list)
print(result)  # Output: Counter({3: 3, 2: 2, 4: 2, 1: 2, 5: 2, 6: 1})
```

```python
from collections import Counter
import math

def top_n_percent(counter, percent=10):
    """
    Returns the top N% of elements from a Counter.
    
    :param counter: Counter object with value counts.
    :param percent: Percentage of elements to keep (default 10%).
    :return: Dictionary with the top N% elements.
    """
    if not counter:
        return {}

    sorted_items = counter.most_common()  # Sort by count (descending)
    top_n = math.ceil(len(sorted_items) * (percent / 100))  # Compute top N elements
    return dict(sorted_items[:top_n])

# Example usage:
counter = Counter({3: 2, 1: 1, 2: 1, 4: 1, 5: 1})
result = top_n_percent(counter, 10)  # Get top 10%

print(result)  
```

```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns

def plot_layer_weights(layer, layer_name, save_plot=False):
    """
    Plots the weights of a specific layer in a PyTorch model as a heatmap.
    
    Parameters:
        layer (torch.nn.Module): The PyTorch layer (e.g., torch.nn.Linear, torch.nn.Conv2d).
        layer_name (str): A text string to label the plot and use in the filename.
        save_plot (bool): If True, saves the plot as a PNG file with the layer_name as part of the filename.
    """
    # Check if the layer has weights
    if not hasattr(layer, 'weight'):
        raise ValueError(f"The provided layer does not have weights.")
    
    # Extract the weights
    weights = layer.weight.data.cpu().numpy()
    
    # Plot the weights as a heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(weights, cmap="coolwarm", annot=False, cbar=True)
    plt.title(f'Weight Matrix of {layer_name}')
    plt.xlabel('Neuron Index')
    plt.ylabel('Input Index')
    
    if save_plot:
        # Save the plot with the layer_name as part of the filename
        filename = f"{layer_name}_weights_heatmap.png"
        plt.savefig(filename, bbox_inches='tight', dpi=300)
        print(f"Plot saved as {filename}")
    
    plt.show()

# Example usage:
model = torch.nn.Sequential(
    torch.nn.Linear(10, 5),
    torch.nn.ReLU(),
    torch.nn.Linear(5, 2)
)

# Access the second Linear layer directly
second_linear_layer = model[2]
plot_layer_weights(second_linear_layer, 'Second_Linear_Layer', save_plot=True)
```

```python
import torch
import matplotlib.pyplot as plt
from codes.models import * 

def plot_dnn_parameters(model, title):
    layer_names = []
    num_params = []
    
    for name, param in model.named_parameters():
        if param.requires_grad:  # Consider only trainable parameters
            layer_names.append(name)
            num_params.append(param.numel())
    
    total_params = sum(num_params)
    
    plt.figure(figsize=(20, 5))
    bars = plt.bar(range(len(num_params)), num_params, tick_label=layer_names)
    
    for bar, param_count in zip(bars, num_params):
        percentage = (param_count / total_params) * 100
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage:.1f}%',
                 ha='center', va='bottom')
    
    plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment to avoid overlap
    plt.xlabel("Layer")
    plt.ylabel("Number of Parameters")
    plt.title(f"Number of Parameters per Layer in DNN {title}")
    plt.tight_layout()  # Adjust layout to better fit x-axis labels
    plt.show()

# # Example usage with a simple model
# if __name__ == "__main__":
#     model = torch.nn.Sequential(
#         torch.nn.Linear(784, 128),
#         torch.nn.ReLU(),
#         torch.nn.Linear(128, 64),
#         torch.nn.ReLU(),
#         torch.nn.Linear(64, 10)
#     )
import torchvision.models as models
# model1 = resnet8()
# model1 = models.resnet50(pretrained=False)
model1 = models.vgg19(pretrained=False)
```
